# PCA

```python
db.iquery("remove(PCA_MATRIX)")
```

```python
db.iquery("""create array PCA_MATRIX <c:double not null>
          [from_entity_idx=0:4650519:0:100000; to_entity_idx=0:4650519:0:4650520]""")
```

```python
db.iquery("""store(
    substitute(
     redimension(apply(TRANSACTION, c, double(transaction_amount)), <c:double>
          [from_entity_idx=0:4650519:0:100000; to_entity_idx=0:4650519:0:4650520], sum(c) as c),
     build(<val:double>[i=0:0], 0)), PCA_MATRIX)""")
```

```python
db.iquery("remove(PCA_MATRIX_T)")
```

```python
db.iquery("""create array PCA_MATRIX_T <c:double not null>
           [to_entity_idx=0:4650519:0:100000; from_entity_idx=0:4650519:0:4650520]""")
```

```python
db.iquery("store(redimension(PCA_MATRIX, PCA_MATRIX_T, false), PCA_MATRIX_T)")
```

```python
db.iquery("store(build(<v:double NOT NULL>[from_entity_idx=0:4650519:0:4650520],random()),initial)") 
```

```python
db.iquery("store(build(<v:double NOT NULL>[from_entity_idx=0:4650519:0:4650520],1),left)") 
```

```python
db.iquery(""" 
     store(
       project(
         substitute(
           apply(
             aggregate(
               PCA_MATRIX,
               sum(c) as colsum,
               to_entity_idx
             ),
             colmean, colsum/4650520
           ),
           build(<v:double NOT NULL>[i=0:0], 0.0),
           colmean
         ),
         colmean
       ),
       right
     )""")
```

```python
db.iquery("remove(TSVD_RESULT)")
```

```python
db.iquery("store(tsvd(PCA_MATRIX, PCA_MATRIX_T, 3,0.001,10, initial, left, right), TSVD_RESULT)")
```
